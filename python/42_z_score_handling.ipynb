```python
import numpy as np
import pandas as pd
df=pd.read_csv('placement.csv')
df
```

```python
#Trimming
df['cgpa_score']=(df['cgpa']-df['cgpa'].mean())/df['cgpa'].std()
new_df= df[(df['cgpa_score']<3)&(df['cgpa_score']>(-3))]
```

```python
#trimming in quantile
percentile_75= df['cgpa'].quantile(0.75)
percentile_25=df['cgpa'].quantile(0.25)
iqr= percentile_75-percentile_25
ul= percentile_75 + 1.5*iqr
ll=percentile_25 - 1.5*iqr
new1_df= df[(df['cgpa']<ul) & (df['cgpa']>ll)]
new1_df
```

```python
#winsorization
upl= df['cgpa'].quantile(0.99)
lol=df['cgpa'].quantile(0.01)
df2=df[(df['cgpa']<upl)&(df['cgpa']>lol)]
df2
```
